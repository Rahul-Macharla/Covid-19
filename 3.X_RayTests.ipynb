```python
from keras import models
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
model=models.load_model("/content/drive/My Drive/CovidDataset/covid_model.h5")
```

```python
img_dims = (224,224,3)
```

```python
image1 = cv2.imread("/content/drive/My Drive/CovidDataset/Nrm.jpg")
plt.imshow(image1)
```

```python
image1 = cv2.resize(image1,(img_dims[0],img_dims[1]))
image1= image1.astype('float')/ 255.0
image1 = np.array(image1)
image1 = np.expand_dims(image1, axis=0)
result1 = model.predict(image1)
result1[0, :] = result1[0, :] > 0.5 
if (result1 == 0):
  print("COVID POSITIVE")
else:
  print("COVID NEGATIVE")
```
